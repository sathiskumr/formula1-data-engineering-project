### create script for ref_constructor_color_code

```sql
CREATE OR REPLACE TABLE formula1ext.bronze.ref_constructor_color_code (
    constructor_id STRING,
    color_code STRING,

    -- 16px
    logo_16_white_png STRING,
    logo_16_dark_png STRING,
    logo_16_transparent_png STRING,

    -- 32px
    logo_32_white_png STRING,
    logo_32_dark_png STRING,
    logo_32_transparent_png STRING,

    -- 64px
    logo_64_white_png STRING,
    logo_64_dark_png STRING,
    logo_64_transparent_png STRING,

    circular_png STRING,

    -- SVG (optional for now)
    logo_svg STRING
    
)
USING DELTA;

INSERT OVERWRITE formula1ext.bronze.ref_constructor_color_code
SELECT
    constructor_id,
    color_code,

    -- 16px
    CONCAT(base_url, '/16/white/', asset_name, '.png'),
    CONCAT(base_url, '/16/dark/', asset_name, '.png'),
    CONCAT(base_url, '/16/transparent/', asset_name, '.png'),

    -- 32px
    CONCAT(base_url, '/32/white/', asset_name, '.png'),
    CONCAT(base_url, '/32/dark/', asset_name, '.png'),
    CONCAT(base_url, '/32/transparent/', asset_name, '.png'),

    -- 64px
    CONCAT(base_url, '/64/white/', asset_name, '.png'),
    CONCAT(base_url, '/64/dark/', asset_name, '.png'),
    CONCAT(base_url, '/64/transparent/', asset_name, '.png'),

    CONCAT(base_url, '/circular_logos_32/', asset_name, '.png'),

    -- SVG
    CONCAT(svg_url, '/', asset_name, '.svg')

FROM VALUES

-- constructor_id, color_code, asset_name

('alfa', '#900000', 'alfa-romeo'),
('alphatauri', '#2B4562', 'alphatauri'),
('alpine', '#0090FF', 'alpine'),
('aston_martin', '#006F62', 'aston-martin'),
('caterham', '#005030', 'caterham'),
('ferrari', '#DC0000', 'ferrari'),
('force_india', '#FF80C0', 'force-india'),
('haas', '#FFFFFF', 'haas'),
('hrt', '#B22222', 'hrt'),
('lotus_f1', '#CBA135', 'lotus'),
('lotus_racing', '#004225', 'lotus-racing'),
('manor', '#D71920', 'manor'),
('marussia', '#6E0000', 'marussia'),
('mclaren', '#FF8700', 'mclaren'),
('mercedes', '#00D2BE', 'mercedes'),
('racing_point', '#F596C8', 'racing-point'),
('rb', '#6692FF', 'racing-bulls'),
('red_bull', '#0600EF', 'red-bull'),
('renault', '#FFF500', 'renault'),
('sauber', '#9B0000', 'sauber'),
('toro_rosso', '#0032A0', 'toro-rosso'),
('virgin', '#C8102E', 'virgin'),
('williams', '#005AFF', 'williams')

AS t(constructor_id, color_code, asset_name)

CROSS JOIN (
    SELECT
        'https://raw.githubusercontent.com/sathiskumr/formula1-data-engineering-project/main/ref_images/constructors-png' AS base_url,
        'https://raw.githubusercontent.com/sathiskumr/formula1-data-engineering-project/main/ref_images/constructors-svg' AS svg_url
) b;
```

### Create script for formula1ext.bronze.ref_country_image 

```sql
CREATE OR REPLACE TABLE formula1ext.bronze.ref_country_image (
    country STRING,
    flag_32_png STRING,
    flag_64_png STRING,
    flag_128_png STRING

)
USING DELTA;

INSERT OVERWRITE formula1ext.bronze.ref_country_image
SELECT
    country,
    CONCAT(base_url, '/', asset_name, '-32px.png')  AS flag_32_png,
    CONCAT(base_url, '/', asset_name, '-64px.png')  AS flag_64_png,
    CONCAT(base_url, '/', asset_name, '-128px.png') AS flag_128_png

FROM VALUES
('Argentina', 'flag-argentina'),
('Australia', 'flag-australia'),
('Austria', 'flag-austria'),
('Azerbaijan', 'flag-azerbaijan'),
('Bahrain', 'flag-bahrain'),
('Belgium', 'flag-belgium'),
('Brazil', 'flag-brazil'),
('Canada', 'flag-canada'),
('China', 'flag-china'),
('France', 'flag-france'),
('Germany', 'flag-germany'),
('Hungary', 'flag-hungary'),
('India', 'flag-india'),
('Italy', 'flag-italy'),
('Japan', 'flag-japan'),
('Malaysia', 'flag-malaysia'),
('Mexico', 'flag-mexico'),
('Monaco', 'flag-monaco'),
('Morocco', 'flag-morocco'),
('Netherlands', 'flag-netherlands'),
('Portugal', 'flag-portugal'),
('Qatar', 'flag-qatar'),
('Russia', 'flag-russia'),
('Saudi Arabia', 'flag-saudi-arabia'),
('Singapore', 'flag-singapore'),
('South Africa', 'flag-south-africa'),
('Korea', 'flag-south-korea'),
('Spain', 'flag-spain'),
('Sweden', 'flag-sweden'),
('Switzerland', 'flag-switzerland'),
('Turkey', 'flag-trkiye'),
('UAE', 'flag-united-arab-emirates'),
('UK', 'flag-united-kingdom'),
('USA', 'flag-united-states')

AS t(country, asset_name)

CROSS JOIN (
    SELECT
        'https://raw.githubusercontent.com/sathiskumr/formula1-data-engineering-project/main/ref_images/country-flags-png' AS base_url
) b;
```
