```sql
CREATE OR REPLACE TABLE formula1_incr.gold.ref_country_image (
    country STRING,
    country_code STRING,    -- ISO 3166-1 alpha-3 country code
    flag_32_png STRING,
    flag_64_png STRING,
    flag_128_png STRING
)
USING DELTA;

INSERT OVERWRITE formula1_incr.gold.ref_country_image
SELECT
    country,
    country_code,
    CONCAT(base_url, '/', asset_name, '-32px.png')  AS flag_32_png,
    CONCAT(base_url, '/', asset_name, '-64px.png')  AS flag_64_png,
    CONCAT(base_url, '/', asset_name, '-128px.png') AS flag_128_png

FROM VALUES
('Argentina', 'ARG', 'flag-argentina'),
('Australia', 'AUS', 'flag-australia'),
('Austria', 'AUT', 'flag-austria'),
('Azerbaijan', 'AZE', 'flag-azerbaijan'),
('Bahrain', 'BHR', 'flag-bahrain'),
('Belgium', 'BEL', 'flag-belgium'),
('Brazil', 'BRA', 'flag-brazil'),
('Canada', 'CAN', 'flag-canada'),
('China', 'CHN', 'flag-china'),
('France', 'FRA', 'flag-france'),
('Germany', 'DEU', 'flag-germany'),
('Hungary', 'HUN', 'flag-hungary'),
('India', 'IND', 'flag-india'),
('Italy', 'ITA', 'flag-italy'),
('Japan', 'JPN', 'flag-japan'),
('Malaysia', 'MYS', 'flag-malaysia'),
('Mexico', 'MEX', 'flag-mexico'),
('Monaco', 'MCO', 'flag-monaco'),
('Morocco', 'MAR', 'flag-morocco'),
('Netherlands', 'NLD', 'flag-netherlands'),
('Portugal', 'PRT', 'flag-portugal'),
('Qatar', 'QAT', 'flag-qatar'),
('Russia', 'RUS', 'flag-russia'),
('Saudi Arabia', 'SAU', 'flag-saudi-arabia'),
('Singapore', 'SGP', 'flag-singapore'),
('South Africa', 'ZAF', 'flag-south-africa'),
('Korea', 'KOR', 'flag-south-korea'),
('Spain', 'ESP', 'flag-spain'),
('Sweden', 'SWE', 'flag-sweden'),
('Switzerland', 'CHE', 'flag-switzerland'),
('Turkey', 'TUR', 'flag-trkiye'),
('UAE', 'ARE', 'flag-united-arab-emirates'),
('UK', 'GBR', 'flag-united-kingdom'),
('USA', 'USA', 'flag-united-states')

AS t(country, country_code, asset_name)
CROSS JOIN (
    SELECT
        'https://raw.githubusercontent.com/sathiskumr/formula1-data-engineering-project/main/ref_images/country-flags-png' AS base_url
) b;
```
